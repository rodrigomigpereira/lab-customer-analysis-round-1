```python
## Read the three files into python as dataframes
df1 = pd.read_csv("lab-customer-analysis-round-1-master/lab-customer-analysis-round-1-master/files_for_lab/csv_files/file1.csv")
df2 = pd.read_csv("lab-customer-analysis-round-1-master/lab-customer-analysis-round-1-master/files_for_lab/csv_files/file2.csv")
df3 = pd.read_csv("lab-customer-analysis-round-1-master/lab-customer-analysis-round-1-master/files_for_lab/csv_files/file3.csv")
```

```python
## Show the DataFrame's shape
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

```python
## Standardize header names
    
    # column "ST" of df1 and df2 should be changed to "State" 
    # column "GENDER' of df1 and df2 should be changed to "Gender"
    # organize the order of the columns. So the three df match each other
```

```python
df1_adjusted = df1.rename(columns={"ST": "State","GENDER": "Gender"})
df2_adjusted = df2.rename(columns={"ST": "State","GENDER": "Gender"})
df3_adjusted = df3
```

```python
columns1 = list(df1_adjusted .columns.values)
columns2 = list(df2_adjusted .columns.values)
columns3 = list(df3_adjusted .columns.values)

display(columns1)
display(columns2)
display(columns3)
```

```python
# Define the new order of columns for each DataFrame
df1_adjusted_new_oder_columns = ['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value','Income', 'Monthly Premium Auto', 'Number of Open Complaints','Policy Type', 'Vehicle Class', 'Total Claim Amount']
df2_adjusted_new_oder_columns = ['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value','Income', 'Monthly Premium Auto', 'Number of Open Complaints','Policy Type', 'Vehicle Class', 'Total Claim Amount']
df3_adjusted_new_oder_columns = ['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value','Income', 'Monthly Premium Auto', 'Number of Open Complaints','Policy Type', 'Vehicle Class', 'Total Claim Amount']
```

```python
# Rearrange the columns in each DataFrame
df1 = df1_adjusted[df1_adjusted_new_oder_columns]
df2 = df2_adjusted[df2_adjusted_new_oder_columns]
df3 = df3_adjusted[df3_adjusted_new_oder_columns]
```

```python
columns1 = list(df1.columns.values)
columns2 = list(df2.columns.values)
columns3 = list(df3.columns.values)

display(columns1)
display(columns2)
display(columns3)
```

```python
## Concatenate the three dataframes

df_concatenated = pd.concat([df1, df2, df3], axis=0) # In pandas, axis = 0 refers to the horizontal axis or rows and axis = 1 refers to the vertical axis or columns. 
display(df_concatenated)
```

```python
## Perform the data cleaning operations mentioned
     # Delete the column education and the number of open complaints from the dataframe.
     # Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.
     # Check for duplicate rows in the data and remove if any.
     # Filter out the data for customers who have an income of 0 or less.
```

```python
df_concatenated = df_concatenated.drop(["Education", "Number of Open Complaints"], axis = 1) # Whether to drop labels from the index (0 or ‘index’) or columns (1 or ‘columns’).
df_concatenated
```

```python
df_concatenated['Customer Lifetime Value'] = df_concatenated['Customer Lifetime Value'].str.replace('%', '').astype(float) * 100
df_concatenated
```

```python
df_concatenated = df_concatenated.drop_duplicates() # by default, it removes duplicate rows based on all columns.
df_concatenated
```
